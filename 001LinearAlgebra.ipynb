```python
# Import NumPy
import numpy as np
```

```python
# A vector can be represented as a NumPy array
# Create a one-dimensional vector
vec = np.array([5,1,3])

print(vec)
```

```python
# Create a two-dimensional vector
vec = np.array([[5,1,3],[2,1,4]])

print(vec)
```

```python
import matplotlib.pyplot as plt
```

```python
vec1 = np.array([1,5])
vec2 = np.array([4,-2])

x_vec1 = vec1[0]
y_vec1 = vec1[1]


x_vec2 = vec2[0]
y_vec2 = vec2[1]

vec = np.array([[x_vec1, x_vec2], [y_vec1, y_vec2]])

origin = np.zeros(vec.shape) # origin point

plt.figure(figsize=(6,6))
plt.quiver(*origin, *vec, color=['r','g','b'], scale=1, units='xy')

plt.grid()
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.gca().set_aspect('equal')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(6,6))
ax = fig.gca(projection='3d')

# Create a three-dimensional vector
# The first vector will be [7 7 7]
# The second vector will be [-5 2 2]
vec = np.array([[7,-5],[7,2],[7,2]])
origin = np.zeros(vec.shape) # origin point

ax.quiver(*origin, *vec)

plt.grid()
ax.set_zlim3d(-10, 10)
ax.set_ylim3d(-10, 10)
ax.set_xlim3d(-10, 10)  
plt.show()
```

```python
# Indexing one-dimensional arrays

a = np.array([1, 2, 3, 4, 5, 6])

print("a[0]:", a[0])
print("a[5]:", a[5])
print("a[:3]:", a[:3])
print("a[1:3]:", a[1:3])
print("a[1:-2]:", a[1:-2])
```

```python
# Indexing two-dimensional arrays

a = np.array([[1, 2, 3], [4, 5, 6]])

print("a.shape:", a.shape)

print("a[0]:", a[0])
print("a[1]:", a[1])

print("a[:,0]:", a[:,0])
print("a[:,1]:", a[:,1])
print("a[:,2]:", a[:,2])
#
print("a[0,0]:", a[0,0])
print("a[2,2]:", a[1,2])
```

```python
# Creating a zero vector using zeros function

a = np.zeros(4)
print(a)
```

```python
e1 = np.array([[1],[0],[0]])
e2 = np.array([[0],[1],[0]])
e3 = np.array([[0],[0],[1]])
```

```python
# Creating a ones vector using ones function

a = np.ones([7])
b = np.ones([3,3])

print("a:", a)
print("b:", b)
```

```python
# Vector addition

vec1 = np.array([[0],[7],[3]])
vec2 = np.array([[1],[2],[0]])

res = vec1 + vec2
print(res)
```

```python
# Vector-Scalar Multiplication

# The result will be calculated in the following way
# res = s*vec1[0] + s*vec1[1] + s*vec1[2]

s = 3
vec1 = np.array([5,1,3])

res = s * vec1

print(res)
print(type(res))
```
